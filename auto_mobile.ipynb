```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv('Automobile.csv')
auto.head()
```

```python
sns.distplot(auto['normalized_losses'])
```

```python
sns.distplot(auto['city_mpg'], kde=False, rug=True);
```

```python
sns.jointplot(x='engine_size',y='horsepower',data=auto)
```

```python
sns.jointplot(x='engine_size',y='horsepower',data=auto, kind="hex")
```

```python
sns.jointplot(x='engine_size',y='horsepower',data=auto, kind="kde")
```

```python
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']])
```

```python
sns.stripplot(x='fuel_type', y='horsepower',data=auto)
```

```python
sns.swarmplot(x='fuel_type', y='horsepower',data=auto)
```

```python
sns.boxplot(x='number_of_doors',y='horsepower', hue='fuel_type',data=auto)
```

```python
sns.barplot(x='body_style',y='horsepower',data=auto,hue='engine_location')
```

```python
sns.countplot(x='body_style',data=auto)
```

```python
sns.pointplot(x='fuel_system',y='horsepower',hue='number_of_doors',data=auto)
```

```python
sns.factorplot(x="fuel_type", 
               y="horsepower", 
               hue="number_of_doors", 
               col="engine_location", 
               data=auto, 
               kind="swarm")
```

```python
#function to linearrigration models
sns.lmplot(x="horsepower", y="peak_rpm", data=auto)
```

```python
sns.lmplot(x="horsepower", y="peak_rpm",hue="fuel_type", data=auto)
```
